#Program to Demonstrate Basic Operations on Data Structures (List) 

```python
# Basic operations on list 

lst = [] 

#insertion

lst.append(10)
lst.append(20)
lst.append(30)
lst.append(40)

print("List after insertion:", lst)

print("===========================================")
#Traversal
print("Traversal of list:")
for item in lst:
    print(item,end=' ')


print("===========================================")
#Searching
key = 20
if key in lst:
    print(f"\n{key} found in the list.")
else:
    print(f"\n{key} not found in the list.")


print("===========================================")
#Deletion
lst.remove(20) 
print("List after deletion of 20:", lst)


print("===========================================")
#length
print("Length of the list:", len(lst))
```
