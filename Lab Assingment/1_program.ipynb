#Program to Insert and Delete an Element at a Given Position in a List

```python
lst = [10, 20, 30, 40, 50] 

print("Original list:", lst)

#insert 

pos =int(input("Enter the position to insert(0-based index): "))
element = int(input("Enter the value to insert: "))

print(f"inserted  element:{element} at position :{pos}")

if pos >=0 and pos <= len(lst):
    lst.insert(pos, element)
    print("list after insertion:", lst)
else:
    print("invalid postion for insettion")

print("==================================================================")

#delete
pos = int(input("Enter the position to delete(0-based index): "))

if pos >=0 and pos < len(lst):
    deleted_element = lst.pop(pos)
    print("Deleted element:", deleted_element)
    print("List after deletion:", lst)
else:
    print("Invalid position for deletion")
```
