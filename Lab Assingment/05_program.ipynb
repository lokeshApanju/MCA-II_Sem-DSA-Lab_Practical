#Program to Represent Two-Dimensional Array Using Python List 

```python
rows = int(input("Enter number of rows: "))
cols = int(input("Enter number of columns: "))

matrix = []

for i in range(rows):
    row = []  # Create a new row for each iteration
    for j in range(cols):
        element = int(input(f"Enter element at position ({i},{j}): "))
        row.append(element)
    matrix.append(row)  # Append the completed row to the matrix

print("2D Array (Matrix):", matrix)

# Print the matrix in grid format
for i in range(rows):
    for j in range(cols):
        print(matrix[i][j], end=" ")
    print()
```
